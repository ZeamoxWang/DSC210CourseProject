## Ising model

Ising model named after Dr. Ernst Ising is a statistical model used to define behaviour of ferromagnetic materials under different temperatures. Ferromagnetic materials refer to materials similar to Iron, which display magnetism.
Consider the atomic or molecular structure of a ferromagnet. In one dimension, we can assume the structure to be a linear chain, and every molecule or atom at each lattice site $i$. For a 2D case, we can assume a rectangular mesh of dimensions $L \times M$, with the molecule or atom at each site $(i,j)$. For simplicity a square mesh of dimensions $ L \times L$ can be assumed. 

Each lattice site consisting of a molecule or atom at lattice site $(i,j)$ is assigned an arbitrary value called the magnetic moment or spin $\sigma_{i,j}$. The spin can take values of $\pm 1$ which correspond to the 'axis of spin'. For a ferromagnetic subtance at lower temperature, almost all of the atoms or molecules have the same spin configuration, i.e. the substance shows magnetic property. 

Ising model aims to find a critical temperature $T_c$ above which the substance loses its magnetic properties, which is interpreted as the lattice configuration transitioning from an almost aligned state to complete random. The net magnetism reduces to zero or in other words every spin atom is paired with an opposite spin. This is termed Phase transition, and can be expressed in terms of relating the total magnetization of the system with the temperature.

### Model definition for a 1D lattice chain

Consider a 1D chain with $N$ lattice sites, with $\nu_1$ positive and $\nu_2$ negative values of spin $\sigma$. 
$$ N = \nu_1 + \nu_2 $$ 

For any two adjacent sites $i$ and $j$, assume there is an interaction energy described by $J_{ij}$. Assume an external magnetic field $B$ is applied to the system. The total energy of the system in a state $\mu$ is descibed ny the hamiltonian $H$ such as 
$$ H = E_{\mu} = E_0 + E_1$$
where $E_0$ is the intermolecular interation energy and $E_1$ is the energy of the system in presence of the magnetic field.
$$ E_{\mu} = - \sum_{(i,j)} J_{ij} \sigma_i \sigma_j  - B \sum_i \sigma_i $$

The probability of the system for being in state configuration $\mu$ is described by the Boltzmann distribution, i.e. 
$$ p_{\mu} = \frac{e^{-\beta E_{\mu}}}{Z}  $$
where $\beta = 1/(k_B.T) $ is called the inverse temperature ( $k_B$ is the Boltzmann constant and $T$ is the temperature of the system). And 
$$ Z = \sum_{\mu} e^{-\beta E_{\mu}} $$ 
is called the partition function. This involves sum of all possible states of the system, which is generally infeasible. Even for a 1D chain with $N$ sites, total number of states is $2^N$. Rewriting the expression for $Z$, and assuming $K_1 = \beta J$ and $H = \beta B$, 
$$ Z = \sum_{\mu} exp(K_1 \sum_i \sigma_i \sigma_{i+1}) \ exp(H \sum_i \sigma_i) $$

$Z$ is of particular interest since we can estimate the free energy per site and in turn estimate the total Magnetization of the system using $Z$. 


**Connection to Linear Algebra**

Rewriting $Z$ has the advantage that now we can formulate $Z$ in terms of a state transfer matrix $V$. Solving:
$$ Z = \sum_{\mu} [exp(K_1 \sigma_1 \sigma_2)][exp(0.5 H(\sigma_1 + \sigma_2))] \cdots [exp(K_1 \sigma_N \sigma_1)][exp(0.5 H(\sigma_N + \sigma_1))]  $$

We can write $[exp(K_1 \sigma_i \sigma_{i+1})]$ as a matrix $V_1$ :
$$ V_1 = \left[ \begin{array}{cc} e^{K_1} & e^{-K_1} \\ e^{-K_1} & e^{K_1} \end{array} \right] $$

and similarly $[exp(0.5 H(\sigma_i + \sigma_{i+1}))]$ can be written as $V_2$ :
$$ V_2 = \left[ \begin{array}{cc} e^{H} & 0 \\ 0 & e^{-H} \end{array} \right] $$

These matrices individually capture the different energies of the system. Therefore, the transfer matrix 
$$ V = V_1 + V_2 = \left[ \begin{array}{cc} e^{K_1 + H} & e^{-K_1} \\ e^{- K_1}& e^{K_1 - H} \end{array} \right]$$

Substituting the values in the equation for $Z$, it can be shown that:
$$\begin{gather*}
Z = \sum_{\sigma_1} V(\sigma_1 \sigma_1)^N\\
Z = Tr(V^N)
\end{gather*}$$
Therefore, the partition function $Z$ can be expressed as the trace of the transfer matrix $V$ multiplied $N$ times. Intuitively this means the partition function is only dependent only on the interaction constants. 

Since, $V$ is real and symmetric, and we can diagonalize $V$ as
$$ V = PDP^{-1}$$

where matrix D is a diagonal matrix with eigne values $\lambda_1$ and $\lambda_2$, and P is a matrix containing the corressponding Eigen vectors $\vec{x_1}$ and $\vec{x_2}$ of V

$$ D = \left[\begin{array}{cc} \lambda_1 & 0 \\ 0 & \lambda_2 \end{array} \right] \quad P =[\vec{x_1}, \vec{x_2}]$$

$$ V^N  = PDP^{-1} PDP^{-1} \cdots PDP^{-1} = PD^NP^{-1}$$

by the property of traces, 
$$ \begin{gather*} Tr(V^N) = Tr(PD^NP^{-1}) = Tr(D^N) \\
Z = Tr(D^N) = \lambda_1^N + \lambda_2^N
\end{gather*}$$

Let $\lambda_1 > \lambda_2$:
$$Z = \lambda_1^N + \lambda_2^N = \lambda_1\bigg(1 + \bigg(\frac{\lambda_2}{\lambda_1}\bigg)^N\bigg)$$

As $N \rightarrow \infty $ , $\lambda_2$ becomes negligible and approaches $0$, hence $Z$:
$$ Z = \lambda_1^N $$

Solving for Eigen values of $V$ we get:
$$ \lambda = e^{K_1} \cosh{H} \pm (e^{2K_1} \sinh^2{H} + e^{-2K_1}) ^{\frac{1}{2}}$$

Steps to estimate total Magnetization $M$ for a 1D case:
1)  Estimate the transfer matrix $V$ based on the interaction energies
2)  Find the largest eigen value of $V$ = $\lambda_1$
3)  Find $Z$ as the $Nth$ power of the largest eigen value.
4)  Get the free energy of system as:
$$f(B,T) = -\lim_{N \to \infty} \frac{1}{\beta N} \ln{Z(B)} $$
$$f(H,T)= -k_B T \ln{\lambda_1}  = -k_BT[e^{K_1} \cosh{H} + (e^{2K_1} \sinh^2{H} + e^{-2K_1}) ^{\frac{1}{2}}]$$
5)  Estimate $M$ as 
$$ M = -\frac{\partial}{\partial B} f(H,T) $$


After finding and substituting all the values: 
$$ M = \frac{e^{K_1}\sinh{H}}{\sqrt{e^{2K_1}\sinh{H} + e^{-2K_1}}} $$

For the 1D case we observe that there is no phase transition at any temperature. The absence of a phase transition in the is reflected in the behavior of the computed total magnetization $M$, which does not show a sharp change or tend to a non-zero value as the temperature varies. $M$ is an analytical function for real vales of $K_1$ and $H$.  The lack of long-range order in one dimension prevents the development of a well-defined ferromagnetic phase. 1D model is an oversimplication of the problem at hand so its natural to extend the model in higher dimensions for practical purposes.

### Two - Dimensional Ising Model

Assuming a rectangular grid, of dimensions $N \times M$. We consider the periodic boundary condition, where the grid could be assumed to be in a Toroidal shape.

The transfer matrix formulism still works. $V_1$ and $V_2$ can be defined in the same way: 
$$ V_1 = \left[ \begin{array}{cc} e^{K_1} & e^{-K_1} \\ e^{-K_1} & e^{K_1} \end{array} \right] \quad \quad V_2 = \left[ \begin{array}{cc} e^{H} & 0 \\ 0 & e^{-H} \end{array} \right] $$

For 2D case, we need to capture the spin interactions in different dimensions so using the Pauli matrices $\tau_x,\tau_y,\tau_z$
$$ \begin{gather*} \tau_x = \left[ \begin{array}{cc} 0 & 1 \\ 1 & 0 \end{array} \right] \quad \quad \tau_y = \left[ \begin{array}{cc} 0 & -i \\ i & 0 \end{array} \right] \quad \quad \tau_z = \left[ \begin{array}{cc} 1 & 0 \\ 0 & -1 \end{array} \right] \\
V_1 = 1.e^{K_1} + \tau_x e^{-K_1}  \text{  or } \quad V_1 = (2\sinh{2K_1})^{\frac{1}{2}}exp(e^{-2K_1}\tau_x)\\
V_2 = 1.\cosh {H} + \tau_z \sinh{H}\text{  or } \quad V_2 = exp(H\tau_z)
\end{gather*}$$

Now, we can generalize to 2D case. In the rectangular case where $M \neq N$, the bond strengths in the lattice are different for rows and columns - assuming them to be $J_1$ and $J_2$ respectively. In 2D cases, the possible orientations increase to $2^M$ for all the possible orientations of the rows. The matrices need to be summed over these states - 
$$ \begin{gather*} 
V_1^{(2)} = (2\sinh{2K_1})^{\frac{M}{2}}exp(e^{-2K_1}\sum_m\tau_m^x)\\
V_2^{(2)}= exp(H\sum_m\tau_m^z)
\end{gather*}$$

Additionaly we need to consider an extra matrix for the column interaction energy as $V_3$:
$$ V_3^{(2)} = exp(K_2\sum_m\tau_m^z\tau_{m+1}^z)$$

The transition matrix and partition function can be expressed as:
$$ V^{(2)} = (V_2^{(2)}V_3^{(2)})^{\frac{1}{2}} V_1^{(2)} (V_2^{(2)}V_3^{(2)})^{\frac{1}{2}}$$
$$Z = Tr[V^{(2)}]^N$$

In Onsagerâ€™s original solution, he diagonalized $V^{(2)}$ by looking for irreducible representations. Further constraints can be imposed by using the Perron-Frobenius Theorem, which gives us that for a real square matrix with positive entries, we can choose the corresponding eigenvector to have strictly positive components. The final result for the Ising 2D case as presented by Onsager:



$$-\beta f = \ln(2) + \frac{1}{8\pi^2} \int_0^{2\pi}d\theta_1 \int_0^{2\pi}d\theta_2 \ln{[\cosh{(2\beta J_1)}\cosh{(2\beta J_2)} - \sinh{(2\beta J_1)}\cos{(\theta)} -\sinh{(2\beta J_2)}\cos{(\theta_2)}]} $$

Solving this we observe there is a phase transition at the critical temperature $T_c$ as defined by:
$$ \sinh{(\frac{2J_1}{k_BT_c})}\sinh{(\frac{2J_2}{k_BT_c})} = 1 $$
For a square lattice, where $J_1 = J_2$
$$ T_c = \frac{2J}{k_B \ln{(1+\sqrt{2})}} $$

This is the final result for Ising 2D case - however the exact solution for this equation is very tricky and has been even dubbed as *"...one of the most complicated and involved calculations in all of theoretical physics. Needless to say, no one has ever been able to find an analytic solution of the Ising model in more than two dimensions."*

```python
import numpy as np
import matplotlib.pyplot as plt
import numba
import math
from numba import njit
from array2gif import write_gif
from IPython.display import display, Image
```

```python
# helpers to calculate total energy and energy after a flip
@numba.njit
def calculate_total_energy(lattice):
    #input: lattice configuration
    #output: total energy of the lattice assuming periodic boundary
    N = len(lattice)
    energy = 0.0
    # periodic boundary conditions
    for i in range(N):
        for j in range(N):
            left_neighbor = lattice[i][(j - 1) % N] if j > 0 else lattice[i][(N - 1)]
            right_neighbor = lattice[i][(j + 1) % N] if j < N - 1 else lattice[i][0]
            above_neighbor = lattice[(i - 1) % N][j] if i > 0 else lattice[(N - 1)][j]
            below_neighbor = lattice[(i + 1) % N][j] if i < N - 1 else lattice[0][j]

            nb_sum = left_neighbor + right_neighbor + above_neighbor + below_neighbor
            energy += -lattice[i][j] * nb_sum

    return energy

@numba.njit
def calculate_energy_change(lattice, x, y, lat_f=None):
    N = len(lattice)
    energy = 0.0
    spin = lat_f if lat_f is not None else lattice[x, y]

    if x > 0:
        energy += -spin * lattice[x - 1, y]
    if x < N - 1:
        energy += -spin * lattice[x + 1, y]
    if y > 0:
        energy += -spin * lattice[x, y - 1]
    if y < N - 1:
        energy += -spin * lattice[x, y + 1]

    return energy
```

```python
# Metropolis Hastings - returns the total magnetization, total energy and the lattice state at each step
@numba.njit
def metropolis(lattice, trials, BJ, energy):
    #lattice : the initial state of the lattice (2d ising)
    #trials: no. of steps for the MCMC
    #BJ: \beta * J or the inverse of the temperature
    #energy: initial energy of the lattice
    
    # 1. init lattice
    N = len(lattice)
    lattice = lattice.copy()
    net_spins = np.zeros(trials-1)
    net_energy = np.zeros(trials-1)
    states = np.zeros((trials-1,N,N))
    for t in range(0, trials-1):
        # 2. pick a random point in the lattice and flip the spin
        x = np.random.randint(0,N)
        y = np.random.randint(0,N)
        lat_i = lattice[x,y]    #initial spin
        lat_f = lat_i * -1      #flip
        
        # compute change in energy
        E_i = calculate_energy_change(lattice, x, y)
        E_f = calculate_energy_change(lattice, x, y, lat_f)
        
        dE = E_f-E_i
        
        # 3 if CONDITION (based on energy change) then accept
        if (dE > 0)*(np.random.random() < np.exp(-BJ*dE)):
            lattice[x,y] = lat_f
            energy += dE
        elif dE <= 0:
            lattice[x,y] = lat_f
            energy += dE
        # 4. else stay in the same state  
        net_spins[t] = lattice.sum()
        net_energy[t] = energy  
        states[t] = lattice  
    return net_spins, net_energy, states
```

```python
# a slightly modified version of the algo defined above, with difference in energy calculations
# and an extra loop of NxN to ensure faster and better convergence
@numba.njit
def metropolis_mod(lattice, trials, BJ, energy):
    N = len(lattice)
    lattice = lattice.copy()
    net_spins = np.zeros(trials-1)
    net_energy = np.zeros(trials-1)
    spins = lattice.sum()
    e = energy
    # outer MH loop
    for t in range(0, trials-1):
        # inner reinit loop
        for n in range(N*N):
            x = np.random.randint(0,N)
            y = np.random.randint(0,N)
            lat_i = lattice[x,y] #initial spin
            lat_f = lat_i * -1 #proposed spin flip

            E_i = calculate_energy_change(lattice, x, y)
            E_f = calculate_energy_change(lattice, x, y, lat_f)
            
            dE = E_f-E_i
        
            if (dE > 0)*(np.random.random() < np.exp(-BJ*dE)):
                lattice[x,y] = lat_f
                e += dE
                spins += 2*lat_f
            elif dE <= 0:
                lattice[x,y] = lat_f
                e += dE
                spins += 2*lat_f
            
        net_spins[t] = spins
        net_energy[t] = e
    return net_spins, net_energy
```

```python
#helpers to visualize the output in a gif
def to_two_color(lattice):
    blue = np.ones(lattice.shape, dtype=np.int) * 255 
    red = np.zeros(lattice.shape, dtype=np.int)
    red[lattice < 0] = 255 
    green = red 
    return np.array([red, green, blue])


def output_to_clip(dataset, fps=20, sample_rate = 1000):
    print("Frames: {}".format(len(dataset)))
    sample = []
    for i in range(len(dataset)):
        if i%sample_rate == 0:
            sample.append(to_two_color(dataset[i]))
    return sample
```

```python
# helps to run MH and visualize first and final states of the lattice
def run_and_visualize(L, init_prob, temp):
    N = L
    warmup = 1000
    trials = 100000
    prob = init_prob
    init_random = np.random.random((N,N))
    lattice = np.zeros((N, N))
    lattice[init_random>=prob] = 1
    lattice[init_random<prob] = -1
    
    net_mag, net_energy, final_state = metropolis(lattice, 
                                                  trials, 
                                                  1/temp, 
                                                  calculate_total_energy(lattice))
    #visualize start and final states of lattice
    fig, axes = plt.subplots(1, 2, figsize=(12,4))
    axes[0].imshow(lattice)
    axes[0].set_title(f"Initial State, {(1-init_prob)*100:.1f} % parallel spins")
    axes[1].imshow(final_state[-1])
    axes[1].set_title("Final equilibrium state")
    plt.show()
    
    return net_mag, net_energy, final_state
```

```python
# @numba.njit
def re_init(lattice, bj):
    N = len(lattice)
    lattice = lattice.copy()
    for n in range(N*N):
        x = np.random.randint(0,N)
        y = np.random.randint(0,N)
        lat_i = lattice[x,y]
        lat_f = lat_i * -1
        
        E_i = calculate_energy_change(lattice, x, y)
        E_f = calculate_energy_change(lattice, x, y, lat_f)
        
        dE = E_f-E_i
        if (dE > 0)*(np.random.random() < np.exp(-bj*dE)):
            lattice[x,y] = lat_f
        elif dE <= 0:
            lattice[x,y] = lat_f
    return lattice

#helper for experiment 1
def exp_1(lattice, BJs, trials, warmup):
    
    E_std = np.zeros(len(BJs))
    
    M_absmeans = np.zeros(len(BJs))
    
    for i, bj in enumerate(BJs):
        spins, energies, _ = metropolis(lattice, trials, bj, calculate_total_energy(lattice))

        E_std[i] = energies[-warmup:].std()
        
        M_absmeans[i] = ((spins[-warmup:]).mean())
        
    return [E_std], [M_absmeans]

#helper for experiment 2
def exp_2(lattice, BJs, trials, warmup):

    BJs_flip = np.flip(BJs)
    E_means = np.zeros(len(BJs))
    E_sqmeans = np.zeros(len(BJs))
    E_std = np.zeros(len(BJs))
    
    M_means = np.zeros(len(BJs))
    M_sqmeans = np.zeros(len(BJs))
    M_std = np.zeros(len(BJs))
    M_absmeans = np.zeros(len(BJs))
    M_4means = np.zeros(len(BJs))
    
    
    for i, bj in enumerate(BJs_flip):
        lattice = re_init(lattice, bj)
        spins, energies = metropolis_mod(lattice, trials, bj, calculate_total_energy(lattice))

        E_means[i] = (energies[-warmup:].mean())
        E_sqmeans[i] = (np.square(energies[-warmup:]).mean())
        E_std[i] = (energies[-warmup:].std())
        
        M_means[i] = (spins[-warmup:].mean())
        M_sqmeans[i] = (np.square(spins[-warmup:]).mean())
        M_std[i] = (spins[-warmup:].std())
        M_absmeans[i] = (np.abs(spins[-warmup:]).mean())
        M_4means[i] = (np.power(spins[-warmup:],4).mean())
        
    return [E_means, E_sqmeans, E_std], [M_means, M_sqmeans, M_absmeans, M_4means, M_std]
```

```python
from scipy.optimize import curve_fit
def powerlaw_susc(T, k=1, gamma=-1, Tc=2.26):
    return k * np.abs(T-Tc)**gamma
```

```python
Ts = 1/BJs
tt = [11,7,5,6]
Ks = np.zeros(len(sizes))
gammas = np.zeros(len(sizes))
TCs = np.zeros(len(sizes))

plt.figure(figsize=(16,15))
for i, xm in enumerate(X_m):
  ax=plt.subplot(3,2,i+1)

  Tmax = np.argmax(xm[2:])
  sol = curve_fit(powerlaw_susc, Ts[Tmax+tt[i]:], xm[Tmax+tt[i]:], maxfev=int(1e6) )[0]
  Ks[i] = sol[0]
  gammas[i] = sol[1]
  TCs[i] = sol[2]

  ax.plot(Ts[Tmax+tt[i]:], powerlaw_susc(Ts[Tmax+tt[i]:], Ks[i], gammas[i], TCs[i]),label='Powerlaw fit')
  ax.plot(Ts[Tmax+tt[i]:], xm[Tmax+tt[i]:],'o',label='points')
  ax.set_xlabel('Temperature')
  ax.set_ylabel('Susceptibility')
  ax.set_title('Lattice size='+str(sizes[i]))
  plt.legend()
  plt.grid()
  print(gammas[i], TCs[i])
print(gammas[1:].mean())
print(TCs[1:].mean())
plt.subplots_adjust(hspace=0.3)  
plt.show()
```
